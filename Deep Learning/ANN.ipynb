````python
import numpy as np
import pandas as pd
import tensorflow as tf
````

````python
dataset=pd.read_csv("/content/drive/MyDrive/Machine Learning A-Z (Codes and Datasets)/Part 8 - Deep Learning/Section 39 - Artificial Neural Networks (ANN)/Python/Churn_Modelling.csv")
x=dataset.iloc[:,3:-1].values
y=dataset.iloc[:,-1].values
````

````python
print(x)
````

````python
print(y)
````

````python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
x[:,2]=le.fit_transform(x[:,2])
````

````python
print(x)
````

````python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
ct=ColumnTransformer(transformers=[('encoder',OneHotEncoder(),[1])],remainder='passthrough') # all remaining columns that were not specified in transformers will be automatically passed through
x=np.array(ct.fit_transform(x))
print(x)
````

````python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=1)
````

````python
#Feature Scaling(Always necessary for deep learning)
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.transform(x_test)
````



```
Building the ANN```



````python
#initializing the ann

ann = tf.keras.models.Sequential()
````

````python
#Adding the input layer and the first hidden layer

ann.add(tf.keras.layers.Dense(units=6,activation='relu'))   #First hidden layer is added
````

````python
#Adding second hidden layer

ann.add(tf.keras.layers.Dense(units=6,activation='relu')) 
````

````python
#Adding output layer


ann.add(tf.keras.layers.Dense(units=1,activation='sigmoid')) 
````

````python
#Training the ann

#Compiling the ann

ann.compile(optimizer='adam' , loss='binary_crossentropy' , metrics=['accuracy'])
````

````python
#Training the ann on training set

ann.fit(x_train,y_train,batch_size=32,epochs=100)
````

````python
#Making the predictions and evaluating the model

print(ann.predict(sc.transform([[1,0,0,600,1,40,3,60000,2,1,1,50000]])) > 0.5)
````

````python
# Predicting the Test set results
y_pred = ann.predict(x_test)
y_pred = (y_pred > 0.5)
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1))
````

````python
# Making the Confusion Matrix
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, y_pred)
print(cm)
accuracy_score(y_test, y_pred)
````
