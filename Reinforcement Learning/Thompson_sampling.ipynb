```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset=pd.read_csv("/content/drive/MyDrive/Machine Learning A-Z (Codes and Datasets)/Part 6 - Reinforcement Learning/Section 33 - Thompson Sampling/Python/Ads_CTR_Optimisation.csv")
```

```python
import random
N = 10000
d = 10
N_1 = [0]*d
N_0 = [0]*d
ads_selected = []
total_reward = 0
for i in range(0,N):
  ad=0
  max_random = 0
  for j in range(0,d):
    beta = random.betavariate(N_1[j] + 1 , N_0[j] + 1)
    if(beta > max_random):
      max_random = beta
      ad=j
  ads_selected.append(ad)
  reward=dataset.values[i,ad]
  if reward == 1:
    N_1[ad] += 1
  else:
    N_0[ad] += 1  
  total_reward = total_reward + reward
```

```python
plt.hist(ads_selected)
plt.title("Histogram")
plt.xlabel("Ads")
plt.ylabel("Number of times each ad was selected")
plt.show()
```
