```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset=pd.read_csv("/content/drive/MyDrive/Machine Learning A-Z (Codes and Datasets)/Part 4 - Clustering/Section 24 - K-Means Clustering/Python/Mall_Customers.csv")
dataset
```

```python
x=dataset.iloc[:,3:].values
#There is no dependent variable y in K-means clustering
```

```python
from sklearn.cluster import KMeans
wcss=[]
for i in range(1,11):
  km=KMeans(n_clusters=i,init="k-means++",random_state=0)
  km.fit(x)
  wcss.append(km.inertia_)
plt.plot(range(1,11),wcss)
plt.title("The Elbow Method")
plt.xlabel("Number of clusters")
plt.ylabel("wcss")
plt.show()  
```

```python
km=KMeans(n_clusters=5,init="k-means++",random_state=0)
y_pred=km.fit_predict(x)
```

```python
print(y_pred)
```

```python
plt.scatter(x[y_pred==0,0],x[y_pred==0,1],s=50,c='red',label='cluster1')
plt.scatter(x[y_pred==1,0],x[y_pred==1,1],s=50,c='blue',label='cluster2')
plt.scatter(x[y_pred==2,0],x[y_pred==2,1],s=50,c='green',label='cluster3')
plt.scatter(x[y_pred==3,0],x[y_pred==3,1],s=50,c='cyan',label='cluster4')
plt.scatter(x[y_pred==4,0],x[y_pred==4,1],s=50,c='magenta',label='cluster5')
plt.scatter(km.cluster_centers_[:,0],km.cluster_centers_[:,1],s=100,c='yellow',label='centroids')
plt.title("clusters of customers")
plt.xlabel("Annulal income")
plt.ylabel("Spending score")
plt.legend()
plt.show() 
```
