```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset=pd.read_csv("/content/drive/MyDrive/Machine Learning A-Z (Codes and Datasets)/Part 2 - Regression/Section 4 - Simple Linear Regression/Python/Salary_Data.csv")
X=dataset.iloc[:,:-1].values
Y=dataset.iloc[:,-1].values
```

```python
print(X)
```

```python
print(Y)
```

```python
#X=X.reshape(1,-1)
#Y=Y.reshape(1,-1)
```

```python
from sklearn.impute import SimpleImputer
sc=SimpleImputer(missing_values=np.nan,strategy="mean")
sc.fit(X)
X=sc.transform(X)
```

```python
print(X)
print(type(X))
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.3,random_state=0)
```

```python
print(X_train)
```

```python
print(X_test)
```

```python
print(Y_train)
```

```python
print(Y_test)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X_train=sc.fit_transform(X_train)
X_test=sc.transform(X_test)
print(X_train)
print(X_test)
```

**Traing SLR on the training set**

```python
from sklearn.linear_model import LinearRegression
slr=LinearRegression()
slr.fit(X_train,Y_train)
```

```python
y_pred = slr.predict(X_test)
```

```python
plt.scatter(X_train,Y_train,color='red')
plt.plot(X_train,slr.predict(X_train),color='blue')
plt.title("Salary vs Experience(Training Set)")
plt.xlabel("Years of Experience")
plt.ylabel("Salary")
plt.show()
```

```python
plt.scatter(X_test,Y_test,color='red')
plt.plot(X_train,slr.predict(X_train),color='blue')
plt.title("Salary vs Experience(Test Set)")
plt.xlabel("Years of Experience")
plt.ylabel("Salary")
plt.show()
```

```python
#Predicting salary for a single input
print(slr.predict([[12]]))
```

```python
#Printing values of coefficient and intercept
print(slr.coef_)
print(slr.intercept_)
```

```python
#Final Equation
print(f"{slr.coef_}X + {slr.intercept_}")
```
